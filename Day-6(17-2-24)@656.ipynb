```python
from sklearn.datasets import make_blobs
#The make_blobs functions is used to generate synthetic datasets
#for clustering and classification tasks.
#This function will create clusters of data points with
#Gaussian distribution
```

```python
#creating random dataset
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
```

```python
data
```

```python
data=make_blobs(n_samples=200,n_features=5,centers=5,cluster_std=1.6,random_state=101)
#n_samples=Total number of points equally divided among clusters.
#n_features= It indicated the number of features(coloumns)
#centers=It determine number of clusters to be generated 
#clusters_std= It selts the standers deviation of the clusters.High value
#makes the clusters to spread out
```

```python
#diff between KNN and K means clustering

#1.)
#KNN is used for classification and regression
# K means is used for clustering problems

#2.)
#KNN is supervised algorithm
#K means is unsurpervised algorithm

#3.)
#To traing KNN,we need a dataset with all the data
#points having class labels
#for training K menas,we no need any such information 

#4.)
#we use KNN to predict the class label or new points
#we use K means to find patterns in a given dataset by grouping datapoints 
#into clusters
```

```python
# # To find the error rate
error_rate=[]
for val in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=val)
    knn.fit(x_train,y_train)
    pred_i=knn.predict(x_test)
    print(pred_i !=y_test)
    error_rate.append(np.mean(pred_i!=y_test))
```
