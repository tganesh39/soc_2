```python
#To handle missing data in data frame by using "np.nan".
import pandas as pd
d={"A":[1,2,np.nan],"B":[5,np.nan,np.nan],"C":[1,2,3],"D":[np.nan,np.nan,np.nan]}
df1=pd.DataFrame(d)
df1
```

```python
#To drop the values with Nan
#dropna() is used

df1.dropna()
```

```python
#dropna(how,thresh,axis,inplace)      syntax -->dropna(parameters)

#how='all' or 'any'

print(df1.dropna(how='all'),"\n\n")   #It will be checking the rows with all Nan values to delete that row

print(df1.dropna(how='any'))  #It will delete the rows which have even single Nan value
```

```python
df1.dropna(how='all',axis=1)
```

```python
df1.isna()  #it returns true if it has Nan
```

```python
#To check the sum of Nan values in each column

df1.isna().sum()
```

```python
#thresh 

#it should have non-Nan value more than or equal to thresh value

df1.dropna(thresh=2)
```

```python
df1.dropna(thresh=2,axis=1)
```

```python
#fillna()

df1.fillna(value=0.0)   #or df1.fillna(0)
```

```python
#printing of average of values in a column and storing it in that Nan values of same column

df1['A'].fillna(value=df1['A'].mean())
```
